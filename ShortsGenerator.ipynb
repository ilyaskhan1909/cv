````python
# -*- coding: utf-8 -*-

# ==============================================================================
# YouTube Shorts Horror Teaser Generator
# ==============================================================================
!pip install -q google-generativeai nest_asyncio edge-tts spacy diffusers transformers accelerate moviepy matplotlib Pillow
!python -m spacy download en_core_web_sm

import re
import os
import torch
import nest_asyncio
import asyncio
import sys # Добавлено для проверки 'google.colab' in sys.modules
import psutil
import gc
import math
import random
import subprocess
from datetime import datetime
from PIL import Image, ImageEnhance
# Используем DiffusionPipeline для SDXL
from diffusers import DiffusionPipeline
from moviepy.editor import *
from moviepy.video.fx.all import crop
import edge_tts
from huggingface_hub import snapshot_download # Может не понадобиться, если LoRA убрана
import numpy as np
import google.generativeai as genai
import json
import traceback
from PIL import Image, ImageEnhance
# import time # Можно раскомментировать для пауз

# Активация nest_asyncio для работы в Colab
nest_asyncio.apply()

# Активация nest_asyncio для работы в Colab
nest_asyncio.apply()

print(f">>> Импорты завершены. Версия NumPy: {np.__version__}") # Добавлено для проверки
print(">>> Если ошибок нет, все готово.")

# ==============================================================================
# Исходный Текст Истории
# ==============================================================================
history_text = """
The gravel crunched under the tires of James’s old truck, a lonely sound in the vast stillness of the valley. Dusk was a bruise of purple and orange bleeding across the jagged peaks that ringed Oakhaven, the village a scattering of dark shapes huddled by the ink-black surface of Whisperwind Lake. He hadn’t been back in fifteen years, not since… not since the accident. Now, his Uncle Thomas was gone, and the cabin, filled with shadows and memories, was his to sort through.
James, barely thirty but feeling a lifetime older, stepped out into the biting late autumn air. Leaves, brittle and brown, skittered across the deserted main street. The air was thin here, sharp in his lungs, a stark contrast to the city smog he’d grown accustomed to. He pulled his collar tighter. Even as a boy, a trainee ranger full of naive confidence, this place had unsettled him. The lake, especially.
The only light emanated from the Hearth & Hollow Inn, its windows glowing like warm amber eyes in the deepening twilight. Inside, a robust log fire crackled and spat, casting dancing shadows on the rough-hewn walls. An elderly woman with eyes as keen as a hawk’s and hair like spun silver watched him approach the counter. Martha, her name tag read. She’d been old even when he was a boy.
“James,” she said, her voice raspy, like dry leaves. It wasn’t a question. “Heard you were coming. Thomas’s cabin… it’s been waiting.”
“Martha,” he acknowledged, forcing a small smile. “Just here to settle things.”
She nodded slowly, her gaze unwavering. “The mountains remember, James. And the lake… she has her moods, especially now, with the cold settling in.” She slid a heavy iron key across the polished wood. “The cabin’s aired. But mind the old ways. The paths get tricky when the mists roll in. Best to be indoors when the midnight bell tolls.”
“Bell?” James frowned. He didn’t remember a bell.
“Some hear it, some don’t,” Martha said, turning to poke the fire. “Just an old habit. Good for keeping folks from wandering where they shouldn’t.” Her words, casual yet weighted, settled uneasily in James’s gut. He remembered the hushed village tales now, the whispers of ‘lake guardians’ appeased with driftwood offerings left on the shore during the full moon, of eerie, bobbing lights – will-o’-the-wisps – that lured the unwary deep into the forest. His uncle, a practical man in most respects, had always seemed to give those tales a wide berth, yet James recalled finding small, carved wooden figures near the lake path by their cabin once.
Old habits die hard in Oakhaven, James thought, a shiver tracing its way down his spine that had nothing to do with the cold. The guilt, a familiar spectre, stirred within him – the memory of the lost hiker, the failed rescue high on these same treacherous slopes years ago. A life he couldn’t save. A weight he still carried.
Sleep was a restless, shallow thing in his uncle’s cabin. The scent of old wood, pipe tobacco, and something else, something faintly aquatic and unsettling, filled the air. Around midnight, drawn by an inexplicable restlessness, James found himself by the edge of Whisperwind Lake.
The water was a sheet of polished obsidian, reflecting the star-dusted sky. Mist, pale and ethereal, coiled like breath over its surface, blurring the line between water and shore. The silence was profound, broken only by the occasional sigh of the wind through the skeletal branches of the surrounding pines. It was a beauty that pricked at his unease, a dangerous allure.
He remembered the hiker, Sarah. Her terrified eyes as the scree gave way. His outstretched hand, just inches from hers. The guilt was a cold stone in his stomach. Had she heard the whispers of this lake, felt its pull?
Then, it came. Faint, hollow, and deeply resonant. A single bell toll, seemingly from across the lake, or perhaps from beneath its still, dark waters. Midnight. Martha’s warning.
James’s breath caught. His heart began a slow, heavy drum against his ribs. He scanned the opposite shore, but there was nothing, only the impenetrable blackness of the forest. And then, he heard it – a soft, almost inaudible singing. It wasn't quite human, more like the wind given a sorrowful voice, a melody that tugged at something primal and forgotten within him.
His gaze, drawn by the unearthly sound, flickered to the edge of the woods where the mist clung thickest. There, weaving between the dark sentinels of the pines, were lights. Pale, bluish-green orbs, dancing with a slow, hypnotic grace. Will-o’-the-wisps. He’d dismissed them as swamp gas, as folklore, during his ranger training. But these… these moved with an intelligence, a purpose. They beckoned.
The narration here should slow, the voice dropping to a hushed, tense whisper.
The air grew colder. The singing, though no louder, felt closer, entwining with the mist, with the silent invitation of the lights. James felt a strange pull, a longing mixed with a profound dread. It was the same feeling he’d had on the mountain ridge with Sarah – that terrifying sense of being on the verge of something irrevocable. He thought of his uncle’s carved figures, of the villagers’ hushed reverence for the lake guardians. Were these their emissaries? Or something else entirely? The rational part of his mind, the ex-forest ranger, screamed at him to turn back, to retreat to the relative safety of the cabin. But another part, the part burdened by guilt and a strange, nascent curiosity, was rooted to the spot, watching the spectral ballet unfold.
One light detached itself from the others, drifting closer to the path that led from the lake into the deeper woods. It pulsed gently, an invitation. The singing seemed to focus, calling his name, or so it felt. Or perhaps it was Sarah’s voice, lost in the wind.
He didn’t consciously decide to follow. His legs simply moved, one foot in front of the other, drawn by that single, beckoning gleam. The mist swirled around him, damp and cloying, obscuring the familiar landmarks of his childhood. The trees loomed like skeletal figures, their branches grasping.
The narration should pick up pace here, sentences shortening, building a staccato, heartbeat-like rhythm.
The light danced ahead. Always just out of reach. Deeper. And deeper. The air grew heavy. He was lost. The ranger’s instincts, once so sharp, were blunted by the unnatural stillness, by the disorienting dance of the will-o’-the-wisp. Panic, cold and sharp, began to prickle. The singing was all around him now, a chorus of sorrowful, alluring voices.
He thought of Martha’s warning. The paths get tricky. He thought of his uncle, who had lived his whole life by this haunted lake. Had Thomas made his peace with these woods, these lights? Or had he simply known when to turn away?
He tripped on an unseen root, sprawling onto the damp earth. The light winked out. Darkness, absolute and suffocating, enveloped him. The singing intensified, swirling like a vortex. He could feel a cold breath on his neck.
“No!” The cry was torn from him, a raw sound of pure terror and defiance. “Sarah!” He wasn’t sure why he called her name. An apology? A plea?
His own voice echoed back, distorted, amplified by the strange acoustics of the fog-bound forest. And in that echo, for a terrifying moment, he thought he heard another voice join his, a faint, sighing response. Then, silence. Or almost.
The distant, hollow bell tolled again. Once. Twice. Its sound cut through the oppressive atmosphere, chilling James to the bone. Did James stumble out of the Oakhaven woods that night, shaken but alive, his own desperate cry having somehow broken the spell, or perhaps mistaken by the woods for one of its own? Or did another light, pale and bluish-green, join the spectral procession by Whisperwind Lake, another soul lured from the path by the mournful singing and the hollow promise of the bell?
The final toll hangs in the autumn air, unanswered. The mist continues to coil, holding its secrets close.
"""

# ==============================================================================
# Глобальная Конфигурация (CONFIG) и Константы
# ==============================================================================
CONFIG = {
    "API_KEYS": {"GOOGLE": None},
    "GEMINI": {
        "MODEL_NAME": 'gemini-1.5-flash',
        "SAFETY_SETTINGS": {
            genai.types.HarmCategory.HARM_CATEGORY_HARASSMENT: genai.types.HarmBlockThreshold.BLOCK_NONE,
            genai.types.HarmCategory.HARM_CATEGORY_HATE_SPEECH: genai.types.HarmBlockThreshold.BLOCK_NONE,
            genai.types.HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: genai.types.HarmBlockThreshold.BLOCK_NONE,
            genai.types.HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: genai.types.HarmBlockThreshold.BLOCK_NONE,
        },
        "TEASER_TARGET_NARRATION_SECONDS_MIN": 25,
        "TEASER_TARGET_NARRATION_SECONDS_MAX": 35,
        "GEMINI_PROMPT_COMMON_INSTRUCTIONS": """
COMMON INSTRUCTIONS FOR ALL TASKS:
1.  **Language:** Respond in English.
2.  **Focus:** Maintain a psychological horror, unsettling, and suspenseful tone.
3.  **Conciseness:** Be direct and to the point. Avoid conversational fluff or self-correction statements.
4.  **Format Adherence:** If JSON output is requested, provide *only* the valid JSON structure. If a single phrase is requested, provide *only* that phrase.
5.  **No External Commentary:** Do not include apologies, explanations of your process, or any text outside the requested output format.
6.  **Avoid Direct Quotes:** Do not directly quote large passages from the input text unless it's absolutely essential for the specific task and very brief. Paraphrase or describe instead.
""",
    },
    "FALLBACKS": {
        "TEASER_SCRIPT": "A creeping dread. Unseen eyes. Reality warps, whispers call. What lurks in the shifting shadows of this place? Some doors are best left unopened. Subscribe for the full nightmare.",
        "IMAGE_DESCRIPTION": "An eerie, unsettling presence in a dark, distorted room.",
    },
    "IMAGE_GENERATION": {
        "STABLE_DIFFUSION_MODEL_NAME": "runwayml/stable-diffusion-v1-5",
        "DEFAULT_STYLE_PREFIX": ("cinematic horror, moody lighting, deep shadows, volumetric fog, film noir grain, unsettling atmosphere, terror, dread"),
        "ADDITIONAL_PROMPT_KEYWORDS": ("ultra realistic photo, sharp focus, high detail, photorealistic, 8K"),
        "NEGATIVE_PROMPT": (
            "ugly, tiling, poorly drawn hands, poorly drawn feet, poorly drawn face, out of frame, extra limbs, disfigured, deformed, body out of frame, blurry, bad anatomy, blurred, watermark, grainy, signature, cut off, draft, low detail, out of focus, poorly drawn, oversaturated, jpeg compression, text, 3D render, game graphics, synthetic, cartoon, anime, stylized, wrong proportions, malformed limbs, extra fingers, bright daylight, cheerful colors, fantasy look, (worst quality, low quality, normal quality:1.4)"),
        "WIDTH": 1024, "HEIGHT": 1024, "INFERENCE_STEPS": 30, "GUIDANCE_SCALE": 7.5,
        "ENHANCE_SHARPNESS_FACTOR": 1.1, "ENHANCE_CONTRAST_FACTOR": 1.15,
        "MIN_IMAGE_FILE_SIZE_BYTES_CHECK": 10240, # 10KB, increased check
    },
    "VIDEO_ASSEMBLY": {
        "SHORTS_WIDTH": 1080, "SHORTS_HEIGHT": 1920, "TARGET_SHORT_DURATION_SECONDS": 58,
        "TARGET_VISUAL_SCENES": 7, # Increased slightly
        "DEFAULT_MIN_CLIP_DURATION_SECONDS": 2.0,
        "FADE_DURATION_SECONDS": 0.25, # Slightly longer fade
        "TEMP_CLIPS_SUBDIR": "temp_video_clips_for_ffmpeg",
    },
    "TTS": {"VOICE": "en-US-JennyNeural", "RATE": "+0%", "PITCH": "-6Hz", "FALLBACK_VOICE": "en-US-GuyNeural"},
    "ANIMATION": {"MIN_ZOOM_FACTOR": 1.08, "MAX_ZOOM_FACTOR": 1.22, "PAN_ZOOM_AMPLITUDE_REDUCTION": 0.4}, # Tuned zoom
    "PATHS": {
        "BASE_OUTPUT_DIR": "/content/shorts_output", "ORIGINAL_STORY_FILENAME": "original_story.txt",
        "TEASER_SCRIPT_FILENAME": "teaser_script.txt", "VISUAL_SCENES_FILENAME": "visual_scenes.json",
        "IMAGE_PROMPTS_FILENAME": "image_prompts_details.json", "SQUARE_IMAGES_SUBDIR": "scene_images_square",
        "AUDIO_SUBDIR": "audio", "FULL_AUDIO_FILENAME": "full_teaser_audio.mp3",
    },
    "EXPORT": {"RENDER_PRESET": 'medium', "RENDER_CRF": "23", "RENDER_FPS": 30, "MAX_THREADS_CPU_CORES": 6},
    "MISC": {"AVERAGE_WORDS_PER_MINUTE_NARRATION": 130} # Slightly adjusted
}
MovementType = {"STATIC": 0, "PAN_RIGHT": 1, "PAN_LEFT": 2, "PAN_UP": 3, "PAN_DOWN": 4, "ZOOM_IN": 5, "ZOOM_OUT": 6}
MOVEMENT_PROBABILITIES = {
    MovementType["STATIC"]: 0.10, MovementType["PAN_RIGHT"]: 0.15, MovementType["PAN_LEFT"]: 0.15,
    MovementType["PAN_UP"]: 0.10, MovementType["PAN_DOWN"]: 0.10, MovementType["ZOOM_IN"]: 0.30, # More zoom_in
    MovementType["ZOOM_OUT"]: 0.10
}
gemini_model_global = None
pipe_base_global = None

# ==============================================================================
# Вспомогательные Функции
# ==============================================================================
def get_ram_usage():
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / (1024 * 1024)

def generate_teaser_text_with_gemini(full_text, model):
    print(f"RAM перед запросом сценария тизера: {get_ram_usage():.2f} MB")
    print("Запрос к Gemini API для генерации СЦЕНАРИЯ ТИЗЕРА...")

    target_min_sec = CONFIG['GEMINI']['TEASER_TARGET_NARRATION_SECONDS_MIN']
    target_max_sec = CONFIG['GEMINI']['TEASER_TARGET_NARRATION_SECONDS_MAX']
    avg_wpm = CONFIG['MISC']['AVERAGE_WORDS_PER_MINUTE_NARRATION']
    target_words_min = int(avg_wpm * (target_min_sec / 60.0))
    target_words_max = int(avg_wpm * (target_max_sec / 60.0))
    common_instructions = CONFIG['GEMINI']['GEMINI_PROMPT_COMMON_INSTRUCTIONS']

    prompt = f"""{common_instructions}
TASK: Analyze the provided psychological horror story. Create a single, coherent, atmospheric, and deeply unsettling teaser script. This script is for narrating a YouTube Short (vertical video format, ~{target_min_sec}-{target_max_sec} seconds, aiming for {target_words_min}-{target_words_max} words).

CRITICAL REQUIREMENTS FOR TEASER SCRIPT:
1.  **Intrigue & Suspense:** Focus on creating mystery, dread, and anticipation. Hint at the horror without revealing too much. The goal is to make viewers *need* to know more.
2.  **Atmosphere:** Evoke a strong sense of unease, the uncanny, or psychological disturbance.
3.  **Narrative Flow:** The script must be a single, continuous piece of narration. No scene numbers, no "Narrator:" prefixes. Just the spoken text.
4.  **Hook:** Start with a compelling hook that immediately grabs attention and establishes a sense of wrongness or mystery.
5.  **Rising Tension:** Build suspense or unease progressively.
6.  **Climax/Question (Implicit or Explicit):** End with a powerful statement, a lingering question, or a disturbing implication that leaves the viewer wanting more. A subtle call to "subscribe for the full story" or similar can be woven in if it feels natural, but prioritize atmospheric horror.
7.  **Originality (Inspired by, not copied):** Do NOT simply extract sentences from the original story. Rephrase, reinterpret, and capture the *essence* and *feeling* of the horror in new, concise wording.
8.  **Pacing:** Imagine this being read slowly, with pauses for dramatic effect.
9.  **Avoid Clichés (unless subverted):** Try for fresh ways to express common horror tropes.
10. **No Spoilers:** Do not reveal key plot twists or the exact nature of the threat if it's meant to be a mystery in the full story.
11. **Target Audience:** Think YouTube Shorts - fast-paced, attention-grabbing, but for a horror niche.
12. **Tone:** Eerie, foreboding, unsettling, questioning reality.
13. **Length Adherence:** Strictly aim for the {target_words_min}-{target_words_max} word count ({target_min_sec}-{target_max_sec} seconds narration).

Full Story Text:
\"\"\"
{full_text}
\"\"\"
Complete Teaser Script Output (English, {target_words_min}-{target_words_max} words, single block of text for narration):
"""
    try:
        response = model.generate_content(prompt, safety_settings=CONFIG['GEMINI']['SAFETY_SETTINGS'])
        print(f"RAM после запроса сценария тизера: {get_ram_usage():.2f} MB")
        if not response.parts:
            print("!!! ВНИМАНИЕ: Ответ Gemini для сценария тизера пуст или заблокирован.")
            print(f"!!! Prompt Feedback: {response.prompt_feedback}")
            return CONFIG['FALLBACKS']['TEASER_SCRIPT']

        teaser_text = response.text.strip()
        # Дополнительная очистка от возможных префиксов, которые Gemini иногда добавляет
        teaser_text = re.sub(r'^(Complete Teaser Script Output|Teaser Script|Script):\s*', '', teaser_text, flags=re.IGNORECASE | re.MULTILINE).strip()
        teaser_text = teaser_text.strip('"\'`')

        if not teaser_text:
            print("!!! ПРЕДУПРЕЖДЕНИЕ: Gemini вернул пустой сценарий тизера после очистки.")
            return CONFIG['FALLBACKS']['TEASER_SCRIPT']

        print(f"Gemini сгенерировал сценарий тизера:\n---\n{teaser_text}\n---")
        word_count = len(teaser_text.split())
        estimated_duration = word_count / (avg_wpm / 60.0) if avg_wpm > 0 else 0
        print(f"(Слов: {word_count}, ~{estimated_duration:.1f} сек озвучки)")
        if not (target_words_min <= word_count <= target_words_max + 10): # Небольшой допуск сверху
             print(f"ПРЕДУПРЕЖДЕНИЕ: Длина тизера ({word_count} слов) вне целевого диапазона ({target_words_min}-{target_words_max}).")
        return teaser_text
    except Exception as e:
        print(f"!!! Ошибка при генерации сценария тизера: {e}"); traceback.print_exc()
        return CONFIG['FALLBACKS']['TEASER_SCRIPT']

def segment_script_into_visual_scenes(full_script_text, model):
    target_scenes_ideal = CONFIG['VIDEO_ASSEMBLY']['TARGET_VISUAL_SCENES']
    print(f"RAM перед сегментацией сценария: {get_ram_usage():.2f} MB")
    print(f"Запрос к Gemini API для сегментации сценария на ~{target_scenes_ideal} ВИЗУАЛЬНЫХ сцен...")

    min_scenes = max(3, target_scenes_ideal - 2)
    max_scenes = target_scenes_ideal + 3
    common_instructions = CONFIG['GEMINI']['GEMINI_PROMPT_COMMON_INSTRUCTIONS']

    prompt = f"""{common_instructions}
TASK: Divide the following complete horror teaser script into {min_scenes} to {max_scenes} logical segments. Each segment should correspond to a distinct visual idea, a shift in narrative focus, a change in tension, or a new atmospheric element. These segments will be used to generate individual images for a video.

CRITICAL REQUIREMENTS FOR SEGMENTATION:
1.  **Output Format:** Provide the output as a JSON array of strings. Each string in the array is one segment of the script. Example: `["Segment one text...", "Segment two text...", "Segment three text..."]`
2.  **Coherence:** Each segment should still make sense on its own, even if brief, and flow logically from the previous one.
3.  **Completeness:** All parts of the original script must be included across the segments. Do not omit text.
4.  **Number of Segments:** Aim for {min_scenes} to {max_scenes} segments. Prioritize logical breaks over strict adherence to this number if it improves visual storytelling.
5.  **Visual Potential:** Think about what could be visually represented for each segment. A segment describing an internal feeling might be short, while a segment describing an action or a visual change might be longer.
6.  **Natural Breaks:** Use natural pauses, sentence endings, or shifts in subject/mood as cues for segmentation.
7.  **Avoid Tiny Segments:** Each segment should be substantial enough to warrant its own visual scene in a short video. Avoid segments that are just a few words unless they are extremely impactful and visually distinct.

Teaser Script to Segment:
\"\"\"
{full_script_text}
\"\"\"
JSON Output (array of strings, {min_scenes}-{max_scenes} segments):
"""
    scenes = None
    try:
        response = model.generate_content(prompt, safety_settings=CONFIG['GEMINI']['SAFETY_SETTINGS'])
        print(f"RAM после ответа Gemini на сегментацию: {get_ram_usage():.2f} MB")
        response_text = response.text.strip()
        # print(f"\nDEBUG_SEGMENTATION: Сырой ответ от Gemini для сегментации:\n{'-'*40}\n{response_text}\n{'-'*40}\n")

        json_str = None
        # Улучшенный поиск JSON блока
        json_match = re.search(r'```json\s*(\[[\s\S]*?\])\s*```', response_text, re.IGNORECASE | re.DOTALL)
        if json_match:
            json_str = json_match.group(1).strip()
            print("DEBUG_SEGMENTATION: JSON извлечен из markdown ```json ... ``` блока.")
        elif response_text.startswith('[') and response_text.endswith(']'):
            json_str = response_text
            print("DEBUG_SEGMENTATION: Ответ Gemini выглядит как прямой JSON массив.")
        else:
            # Попытка найти JSON массив где-то в тексте, если не в ```json``` и не прямой
            potential_json_match = re.search(r'(\[[\s\S]*?\])', response_text)
            if potential_json_match:
                json_str = potential_json_match.group(1).strip()
                print("DEBUG_SEGMENTATION: JSON массив найден внутри более общего текста ответа.")
            else:
                print("DEBUG_SEGMENTATION: Ответ не содержит явного JSON массива.")

        if json_str:
            try:
                # Удаление комментариев и висячих запятых перед парсингом
                json_str = re.sub(r"//.*?\n", "\n", json_str) # Убрать однострочные комментарии
                json_str = re.sub(r"/\*.*?\*/", "", json_str, flags=re.DOTALL) # Убрать многострочные комментарии
                json_str = re.sub(r",\s*([\]}])", r"\1", json_str) # Убрать висячие запятые

                parsed_json = json.loads(json_str)
                if isinstance(parsed_json, list) and all(isinstance(s, str) for s in parsed_json):
                    scenes = parsed_json
                    print("DEBUG_SEGMENTATION: JSON успешно распарсен в список строк.")
                else:
                    print("ПРЕДУПРЕЖДЕНИЕ: Распарсенный JSON не является списком строк.")
                    scenes = None # Сброс, чтобы сработал fallback
            except json.JSONDecodeError as e:
                print(f"!!! Ошибка декодирования JSON при сегментации: {e}\nПроблемная строка (начало): {json_str[:200]}...")
                scenes = None # Сброс, чтобы сработал fallback

        if scenes is None: # Если JSON не найден или невалиден
            print("ПРЕДУПРЕЖДЕНИЕ: Не удалось получить корректный JSON для сегментов. Используется запасная сегментация по предложениям.")
            sentences = re.split(r'(?<=[.!?…])\s+', full_script_text.strip())
            num_sentences = len(sentences)
            sentences_per_segment = max(1, math.ceil(num_sentences / target_scenes_ideal)) if target_scenes_ideal > 0 else 1
            scenes = [" ".join(sentences[i : i + sentences_per_segment]).strip() for i in range(0, num_sentences, sentences_per_segment)]
            scenes = [s for s in scenes if s.strip()] # Убрать пустые строки, если есть
            print(f"Запасная сегментация создала {len(scenes)} сцен.")

        if not scenes: # Если даже запасная сегментация не дала результата
            print("!!! КРИТИЧЕСКАЯ ОШИБКА: Не удалось сегментировать сценарий никаким способом. Используется полный текст как одна сцена.")
            return [full_script_text]

        cleaned_scenes = [s.strip() for s in scenes if s.strip()]
        print(f"Сценарий разделен на {len(cleaned_scenes)} визуальных сцен.")
        return cleaned_scenes

    except Exception as e:
        print(f"!!! КРИТИЧЕСКАЯ ОШИБКА в функции сегментации: {e}"); traceback.print_exc()
        return [full_script_text] # Fallback на одну сцену с полным текстом

def generate_image_prompt_description(scene_chunk_text, model):
    print(f"Запрос к Gemini для описания изображения для сегмента: '{scene_chunk_text[:80]}...'")
    common_instructions = CONFIG['GEMINI']['GEMINI_PROMPT_COMMON_INSTRUCTIONS']
    style_guidance_example = CONFIG['IMAGE_GENERATION']['DEFAULT_STYLE_PREFIX'] # Только для примера в промпте

    prompt_to_gemini = f"""{common_instructions}
TASK: Analyze the following horror script segment. Generate a single, very concise (5-15 words), evocative descriptive phrase. This phrase should capture the *core visual element* or the *dominant mood/emotion* of the segment. It will be used to prompt an image generation model.

CRITICAL REQUIREMENTS FOR DESCRIPTIVE PHRASE:
1.  **Visual Focus:** Describe what might be *seen* or the *feeling* it evokes visually.
2.  **Extreme Brevity:** 5-15 words MAXIMUM. Shorter is often better if impactful.
3.  **No Style Instructions:** Do NOT include stylistic keywords like "cinematic," "photorealistic," "dark," "eerie," etc. These will be added separately. Focus *only* on the *content* and *mood* of the scene. (Example of what NOT to include: "A dark, eerie, cinematic shot of...")
4.  **No Script Repetition:** Do NOT simply copy or rephrase the script segment. Create a *new* description based on it.
5.  **Output:** Provide ONLY the descriptive phrase. No surrounding text, no "Description:", just the phrase itself.

Script Segment:
\"\"\"
{scene_chunk_text}
\"\"\"
Descriptive Phrase (English, 5-15 words, pure visual/mood description):
"""
    try:
        response = model.generate_content(prompt_to_gemini, safety_settings=CONFIG['GEMINI']['SAFETY_SETTINGS'])
        # print(f"DEBUG_IMG_DESC: Сырой ответ Gemini = '{response.text.strip()}'")
        if not response.parts:
            print("!!! ВНИМАНИЕ: Ответ Gemini для описания изображения пуст или заблокирован.")
            print(f"!!! Prompt Feedback: {response.prompt_feedback}")
            return CONFIG['FALLBACKS']['IMAGE_DESCRIPTION']

        generated_description = response.text.strip()
        # Очистка от возможных префиксов
        generated_description = re.sub(r'^(Descriptive Phrase|Description):\s*', '', generated_description, flags=re.IGNORECASE | re.MULTILINE).strip().strip('"\'`.')

        # Удаляем стилистические префиксы, если Gemini их случайно включил
        style_prefix_lower_tokens = [token.strip().lower() for token in CONFIG['IMAGE_GENERATION']['DEFAULT_STYLE_PREFIX'].split(',')]
        desc_tokens = generated_description.lower().split()
        for prefix_token in style_prefix_lower_tokens:
            if generated_description.lower().startswith(prefix_token):
                print(f"  ПРЕДУПРЕЖДЕНИЕ: Gemini включил '{prefix_token}' в начало описания. Удаляем.")
                generated_description = re.sub(f"^{re.escape(prefix_token)}[,\s]*", "", generated_description, flags=re.IGNORECASE).strip()

        if not generated_description:
            print("ПРЕДУПРЕЖДЕНИЕ: Описание изображения оказалось пустым после очистки.")
            return CONFIG['FALLBACKS']['IMAGE_DESCRIPTION']

        print(f"  Сгенерированное описание для изображения: '{generated_description}'")
        return generated_description

    except Exception as e:
        print(f"!!! Ошибка при генерации описания изображения: {e}"); traceback.print_exc()
        return CONFIG['FALLBACKS']['IMAGE_DESCRIPTION']

def enhance_image(image: Image.Image) -> Image.Image:
    s = CONFIG['IMAGE_GENERATION']['ENHANCE_SHARPNESS_FACTOR']
    c = CONFIG['IMAGE_GENERATION']['ENHANCE_CONTRAST_FACTOR']
    enhanced_image = image
    try:
        if s > 1.0:
            enhancer_sharp = ImageEnhance.Sharpness(enhanced_image)
            enhanced_image = enhancer_sharp.enhance(s)
        if c != 1.0: # Контраст может быть и уменьшен
            enhancer_contrast = ImageEnhance.Contrast(enhanced_image)
            enhanced_image = enhancer_contrast.enhance(c)
        return enhanced_image
    except Exception as e:
        print(f"Предупреждение: Не удалось улучшить изображение: {e}")
        return image # Возвращаем оригинал в случае ошибки

def generate_stable_diffusion_image(prompt, out_path, sd_pipe_model):
    neg_prompt = CONFIG['IMAGE_GENERATION']['NEGATIVE_PROMPT']
    width = CONFIG['IMAGE_GENERATION']['WIDTH']
    height = CONFIG['IMAGE_GENERATION']['HEIGHT']
    guidance_scale = CONFIG['IMAGE_GENERATION']['GUIDANCE_SCALE']
    num_steps = CONFIG['IMAGE_GENERATION']['INFERENCE_STEPS']

    print(f"RAM перед генерацией SD изображения: {get_ram_usage():.2f} MB")
    print(f"Генерация SD ({width}x{height}). Промпт (начало): {prompt[:120]}...")

    if not torch.cuda.is_available() and sd_pipe_model.device.type != 'cpu':
        print("ПРЕДУПРЕЖДЕНИЕ: CUDA недоступна, но модель не на CPU. Попытка переместить на CPU...")
        try:
            sd_pipe_model.to("cpu")
        except Exception as move_err:
            print(f"Не удалось переместить модель SD на CPU: {move_err}")

    try:
        # Генерация изображения
        img_output = sd_pipe_model(
            prompt=prompt,
            negative_prompt=neg_prompt,
            width=width,
            height=height,
            guidance_scale=guidance_scale,
            num_inference_steps=num_steps,
            output_type="pil"
        ).images

        if not img_output or not img_output[0]:
            raise ValueError("Модель Stable Diffusion не вернула изображение (img_output is empty or None).")

        img = img_output[0]
        print("  -> SD генерация завершена.")

        enhanced_img = enhance_image(img)
        print(f"RAM после генерации SD и улучшения: {get_ram_usage():.2f} MB")

        enhanced_img.save(out_path)
        print(f"  Изображение сохранено: {out_path}")
        return out_path

    except torch.cuda.OutOfMemoryError as oom:
        print(f"!!! VRAM OutOfMemoryError при генерации SD: {oom}"); traceback.print_exc()
        print("  Попытка освободить память GPU и повторить на CPU, если возможно...")
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        # В этой версии не будем пытаться повторить на CPU, а создадим заглушку.
        # Если OOM происходит даже на CPU, то это более серьезная проблема.
    except Exception as e:
        print(f"!!! Ошибка при генерации изображения Stable Diffusion: {e}"); traceback.print_exc()

    # Fallback: создать черную заглушку, если генерация не удалась
    try:
        print("  Создание черной заглушки из-за ошибки генерации...")
        placeholder_img = Image.new('RGB', (width, height), color='black')
        placeholder_img.save(out_path)
        print(f"  Черная заглушка сохранена: {out_path}")
        return out_path
    except Exception as placeholder_err:
        print(f"!!! Не удалось сохранить черную заглушку: {placeholder_err}")
        return None

async def generate_audio_tts(text_to_speak, output_path):
    voice = CONFIG['TTS']['VOICE']
    rate = CONFIG['TTS']['RATE']
    pitch = CONFIG['TTS']['PITCH']
    fallback_voice = CONFIG['TTS']['FALLBACK_VOICE']

    print(f"RAM перед генерацией TTS: {get_ram_usage():.2f} MB")
    print(f"Генерация аудио ({voice}, Rate:{rate}, Pitch:{pitch}) для: {os.path.basename(output_path)}")

    # Очистка текста для TTS
    clean_text = text_to_speak.strip()
    clean_text = re.sub(r'[“”"‘’`*#_]', '', clean_text) # Удаление специфичных символов
    clean_text = clean_text.replace('—', '-').replace('…', '...') # Замена тире и многоточий
    clean_text = re.sub(r'\s+', ' ', clean_text).strip() # Нормализация пробелов

    if not clean_text:
        print("ПРЕДУПРЕЖДЕНИЕ: Пустой текст для генерации аудио после очистки.")
        return None

    try:
        communicate = edge_tts.Communicate(clean_text, voice, rate=rate, pitch=pitch)
        await communicate.save(output_path)
        print(f"  Аудио успешно сохранено: {os.path.basename(output_path)}")
        return output_path
    except Exception as e:
        print(f"!!! Ошибка TTS с голосом '{voice}': {e}"); traceback.print_exc()
        if voice != fallback_voice:
            print(f"  Попытка генерации с fallback голосом '{fallback_voice}'...")
            try:
                communicate_fallback = edge_tts.Communicate(clean_text, fallback_voice, rate=rate, pitch=pitch)
                await communicate_fallback.save(output_path)
                print(f"  Аудио с fallback голосом успешно сохранено.")
                return output_path
            except Exception as e2:
                print(f"!!! Повторная ошибка TTS с fallback голосом '{fallback_voice}': {e2}")
                return None
        else:
            print("!!! Fallback голос такой же, как основной. Повторная попытка невозможна.")
            return None

def ease_in_out_quad(t_normalized):
    # t_normalized is time from 0.0 to 1.0
    if t_normalized < 0.5:
        return 2 * t_normalized * t_normalized
    return 1 - pow(-2 * t_normalized + 2, 2) / 2

def create_advanced_make_frame(pil_image_captured, base_img_width, base_img_height,
                               visual_duration, movement_effect_type,
                               zoom_factor_start, zoom_factor_end):
    # Захватываем переменные в замыкании
    img_pil = pil_image_captured
    # w_original, h_original = base_img_width, h_original = base_img_height # <--- БЫЛО НЕПРАВИЛЬНО
    w_original = base_img_width   # <--- ИСПРАВЛЕНО
    h_original = base_img_height  # <--- ИСПРАВЛЕНО

    duration_of_effect = visual_duration
    movement = movement_effect_type
    z_start, z_end = zoom_factor_start, zoom_factor_end

    shorts_target_w = CONFIG['VIDEO_ASSEMBLY']['SHORTS_WIDTH']
    shorts_target_h = CONFIG['VIDEO_ASSEMBLY']['SHORTS_HEIGHT']

    static_frame_np_array = None
    if movement == MovementType["STATIC"]:
        try:
            scale_w = shorts_target_w / w_original if w_original > 0 else 1
            scale_h = shorts_target_h / h_original if h_original > 0 else 1
            scale_to_fit = max(scale_w, scale_h)

            target_w_scaled = int(w_original * scale_to_fit)
            target_h_scaled = int(h_original * scale_to_fit)

            if target_w_scaled <=0 or target_h_scaled <=0: raise ValueError("Static frame scaled to zero or negative size.")

            resized_img_pil = img_pil.resize((target_w_scaled, target_h_scaled), Image.Resampling.LANCZOS)

            left = (target_w_scaled - shorts_target_w) / 2
            top = (target_h_scaled - shorts_target_h) / 2
            right = left + shorts_target_w
            bottom = top + shorts_target_h

            left, top = max(0, int(round(left))), max(0, int(round(top)))
            right, bottom = min(target_w_scaled, int(round(right))), min(target_h_scaled, int(round(bottom)))

            cropped_img_pil = resized_img_pil.crop((left, top, right, bottom))

            if cropped_img_pil.size != (shorts_target_w, shorts_target_h):
                cropped_img_pil = cropped_img_pil.resize((shorts_target_w, shorts_target_h), Image.Resampling.LANCZOS)

            static_frame_np_array = np.array(cropped_img_pil)
        except Exception as static_prep_err:
            print(f"!!! Ошибка при подготовке статического кадра: {static_prep_err}")
            static_frame_np_array = np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

    def make_frame_function(time_t):
        if movement == MovementType["STATIC"]:
            return static_frame_np_array if static_frame_np_array is not None else np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        normalized_time = min(max(time_t / duration_of_effect if duration_of_effect > 0 else 0.0, 0.0), 1.0)
        eased_time = ease_in_out_quad(normalized_time)

        current_zoom = z_start + (z_end - z_start) * eased_time

        render_w = int(round(w_original * current_zoom))
        render_h = int(round(h_original * current_zoom))

        if render_w <= 0 or render_h <= 0:
            return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        try:
            resized_pil_image_for_effect = img_pil.resize((render_w, render_h), Image.Resampling.LANCZOS)
        except ValueError:
             return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)
        except Exception as resize_err:
            print(f"Ошибка ресайза в make_frame: {resize_err}")
            return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        max_pan_x = max(0, render_w - shorts_target_w)
        max_pan_y = max(0, render_h - shorts_target_h)

        offset_x = max_pan_x / 2
        offset_y = max_pan_y / 2

        if movement == MovementType["PAN_RIGHT"]:
            offset_x = max_pan_x * eased_time
        elif movement == MovementType["PAN_LEFT"]:
            offset_x = max_pan_x * (1 - eased_time)
        elif movement == MovementType["PAN_DOWN"]:
            offset_y = max_pan_y * eased_time
        elif movement == MovementType["PAN_UP"]:
            offset_y = max_pan_y * (1 - eased_time)

        crop_left = int(round(offset_x))
        crop_top = int(round(offset_y))
        crop_right = crop_left + shorts_target_w
        crop_bottom = crop_top + shorts_target_h

        if crop_right > render_w: crop_right = render_w; crop_left = max(0, render_w - shorts_target_w)
        if crop_bottom > render_h: crop_bottom = render_h; crop_top = max(0, render_h - shorts_target_h)
        if crop_left < 0: crop_left = 0
        if crop_top < 0: crop_top = 0

        if crop_right <= crop_left or crop_bottom <= crop_top:
            return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        try:
            final_cropped_pil_image = resized_pil_image_for_effect.crop((crop_left, crop_top, crop_right, crop_bottom))
        except Exception as crop_err:
            print(f"Ошибка кадрирования в make_frame: {crop_err}")
            return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        if final_cropped_pil_image.size != (shorts_target_w, shorts_target_h):
            try:
                final_cropped_pil_image = final_cropped_pil_image.resize((shorts_target_w, shorts_target_h), Image.Resampling.LANCZOS)
            except Exception as final_resize_err:
                print(f"Ошибка финального ресайза в make_frame: {final_resize_err}")
                return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        return np.array(final_cropped_pil_image)

    return make_frame_function

    def make_frame_function(time_t):
        if movement == MovementType["STATIC"]:
            return static_frame_np_array if static_frame_np_array is not None else np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        # Нормализованное время (0.0 до 1.0) и применение easing
        normalized_time = min(max(time_t / duration_of_effect if duration_of_effect > 0 else 0.0, 0.0), 1.0)
        eased_time = ease_in_out_quad(normalized_time)

        # Текущий зум
        current_zoom = z_start + (z_end - z_start) * eased_time

        # Размеры изображения после зума, но до кадрирования под шортс
        # Изображение сначала масштабируется так, чтобы его высота * зум стала высотой шортса (для вертикального формата)
        # или ширина * зум стала шириной шортса (если бы был горизонтальный).
        # Для вертикального шортса, мы обычно хотим, чтобы квадратное изображение заполнило его по ширине или высоте,
        # а затем было отзумлено.
        # Правильный подход: масштабируем так, чтобы меньшая сторона отзумленного изображения соответствовала
        # меньшей стороне кадра шортса, ИЛИ чтобы большая сторона отзумленного изображения соответствовала большей стороне кадра.
        # Для "cover" эффекта на шортс:
        # Масштабируем оригинальное изображение так, чтобы оно могло покрыть шортс ПОСЛЕ зума.

        # Размер изображения, которое будет кадрироваться
        # Масштаб, чтобы изображение заполнило шортс после применения зума
        # Целевая высота изображения после зума должна быть как минимум shorts_target_h
        # Целевая ширина изображения после зума должна быть как минимум shorts_target_w

        # Рассчитываем размеры отзумленного исходного изображения
        zoomed_w = w_original * current_zoom
        zoomed_h = h_original * current_zoom

        # Масштаб для вписывания отзумленного изображения в кадр шортса с покрытием
        # Это немного отличается от предыдущей логики, где зум применялся к уже подогнанному под размер кадра изображению.
        # Здесь зум применяется к оригинальному размеру, а потом это отзумленное изображение позиционируется.

        # Для простоты, давайте считать, что наше исходное изображение (1024x1024)
        # должно быть отзумлено и затем вписано/обрезано в 1080x1920.

        # Размер изображения, которое будет отрендерено и из которого будет вырезан кадр
        # Это исходное изображение, отмасштабированное текущим зумом
        render_w = int(round(w_original * current_zoom))
        render_h = int(round(h_original * current_zoom))

        if render_w <= 0 or render_h <= 0:
            return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8) # Черный кадр

        try:
            resized_pil_image_for_effect = img_pil.resize((render_w, render_h), Image.Resampling.LANCZOS)
        except ValueError: # e.g. image has zero width or height
             return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)
        except Exception as resize_err:
            print(f"Ошибка ресайза в make_frame: {resize_err}")
            return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        # Координаты для кадрирования из отзумленного/отмасштабированного изображения
        # чтобы получить финальный кадр размером shorts_target_w x shorts_target_h

        # Максимальное смещение по X и Y
        max_pan_x = max(0, render_w - shorts_target_w)
        max_pan_y = max(0, render_h - shorts_target_h)

        # Начальное смещение (для центрирования или для начала панорамирования)
        offset_x = max_pan_x / 2
        offset_y = max_pan_y / 2

        if movement == MovementType["PAN_RIGHT"]:
            offset_x = max_pan_x * eased_time # от 0 до max_pan_x
        elif movement == MovementType["PAN_LEFT"]:
            offset_x = max_pan_x * (1 - eased_time) # от max_pan_x до 0
        elif movement == MovementType["PAN_DOWN"]: # Pan down (изображение движется вверх, мы видим нижнюю часть)
            offset_y = max_pan_y * eased_time
        elif movement == MovementType["PAN_UP"]: # Pan up (изображение движется вниз, мы видим верхнюю часть)
            offset_y = max_pan_y * (1 - eased_time)
        # Для ZOOM_IN и ZOOM_OUT смещение остается центральным

        crop_left = int(round(offset_x))
        crop_top = int(round(offset_y))
        crop_right = crop_left + shorts_target_w
        crop_bottom = crop_top + shorts_target_h

        # Коррекция, если вышли за границы из-за округления или малого зума
        if crop_right > render_w: crop_right = render_w; crop_left = max(0, render_w - shorts_target_w)
        if crop_bottom > render_h: crop_bottom = render_h; crop_top = max(0, render_h - shorts_target_h)
        if crop_left < 0: crop_left = 0
        if crop_top < 0: crop_top = 0

        # Если после коррекций размеры для кадрирования невалидны
        if crop_right <= crop_left or crop_bottom <= crop_top:
             return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        try:
            final_cropped_pil_image = resized_pil_image_for_effect.crop((crop_left, crop_top, crop_right, crop_bottom))
        except Exception as crop_err:
            print(f"Ошибка кадрирования в make_frame: {crop_err}")
            return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        # Убедимся, что финальный кадр точно нужного размера (может быть +-1 пиксель из-за округлений)
        if final_cropped_pil_image.size != (shorts_target_w, shorts_target_h):
            try:
                final_cropped_pil_image = final_cropped_pil_image.resize((shorts_target_w, shorts_target_h), Image.Resampling.LANCZOS)
            except Exception as final_resize_err:
                print(f"Ошибка финального ресайза в make_frame: {final_resize_err}")
                return np.zeros((shorts_target_h, shorts_target_w, 3), dtype=np.uint8)

        return np.array(final_cropped_pil_image)

    return make_frame_function

# ==============================================================================
# Функции Основных Шагов Пайплайна
# ==============================================================================
def setup_environment_and_models():
    global gemini_model_global, pipe_base_global
    print("\n>>> Настройка окружения и загрузка моделей...")
    print(f"RAM в начале setup: {get_ram_usage():.2f} MB")

    # Настройка API ключа Gemini
    api_key_gemini = None
    try:
        from google.colab import userdata
        api_key_gemini = userdata.get('GOOGLE_API_KEY')
        if api_key_gemini:
            print("GOOGLE_API_KEY успешно получен из Colab Secrets.")
        else:
            print("GOOGLE_API_KEY не найден в Colab Secrets. Попытка из переменных окружения...")
            api_key_gemini = os.environ.get('GOOGLE_API_KEY')
    except ImportError: # Не в Colab
        print("Не в среде Colab. Попытка получить GOOGLE_API_KEY из переменных окружения...")
        api_key_gemini = os.environ.get('GOOGLE_API_KEY')
    except Exception as e_colab: # Другая ошибка с userdata
        print(f"Ошибка при доступе к Colab Secrets: {e_colab}. Попытка из переменных окружения...")
        api_key_gemini = os.environ.get('GOOGLE_API_KEY')

    if not api_key_gemini:
        print("GOOGLE_API_KEY не найден ни в Colab Secrets, ни в переменных окружения.")
        try:
            api_key_gemini = input("Пожалуйста, введите ваш GOOGLE_API_KEY: ").strip()
        except EOFError: # Если скрипт запускается неинтерактивно
             print("Неинтерактивный режим, не могу запросить ключ. Завершение.")
             sys.exit("GOOGLE_API_KEY не предоставлен.")
        if not api_key_gemini:
            sys.exit("GOOGLE_API_KEY не был введен. Завершение.")

    CONFIG['API_KEYS']['GOOGLE'] = api_key_gemini
    # Установка переменной окружения, если она еще не установлена или отличается
    if os.environ.get('GOOGLE_API_KEY') != api_key_gemini:
        os.environ['GOOGLE_API_KEY'] = api_key_gemini
        print("GOOGLE_API_KEY установлен в переменные окружения для этой сессии.")

    # Инициализация модели Gemini
    try:
        genai.configure(api_key=CONFIG['API_KEYS']['GOOGLE'])
        model_name_gemini = CONFIG['GEMINI']['MODEL_NAME']
        # Пробный запрос для проверки работоспособности модели и API ключа
        ping_model = genai.GenerativeModel(model_name_gemini)
        ping_model.generate_content("Hello Gemini.", safety_settings=CONFIG['GEMINI']['SAFETY_SETTINGS']) # Простой пинг

        gemini_model_global = genai.GenerativeModel(model_name_gemini) # Сохраняем глобально
        print(f"Модель Gemini '{model_name_gemini}' успешно инициализирована. RAM: {get_ram_usage():.2f} MB")
    except Exception as genai_err:
        print(f"КРИТИЧЕСКАЯ ОШИБКА при инициализации Gemini: {genai_err}"); traceback.print_exc()
        sys.exit("Не удалось инициализировать модель Gemini.")

    # Инициализация модели Stable Diffusion
    sd_model_name = CONFIG['IMAGE_GENERATION']['STABLE_DIFFUSION_MODEL_NAME']
    device_sd = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"\nПопытка загрузки Stable Diffusion модели: '{sd_model_name}' на устройство: {device_sd}...")

    try:
        pipe_base_global = DiffusionPipeline.from_pretrained(
            sd_model_name,
            torch_dtype=torch.float16 if device_sd == "cuda" else torch.float32, # float16 только для CUDA
            use_safetensors=True,
            # low_cpu_mem_usage=True # Может помочь при очень ограниченной CPU RAM, но может замедлить загрузку
        )
        pipe_base_global.to(device_sd)
        print(f"Модель Stable Diffusion '{sd_model_name}' загружена на {device_sd}. RAM: {get_ram_usage():.2f} MB")

        if device_sd == "cuda":
            print("Попытка включить CPU Offloading для Stable Diffusion (если достаточно VRAM)...")
            try:
                # Включаем offloading только если основная модель поместилась в VRAM
                pipe_base_global.enable_model_cpu_offload()
                print(f"CPU Offloading для SD успешно включен. RAM: {get_ram_usage():.2f} MB")
            except Exception as offload_err:
                print(f"Предупреждение: не удалось включить CPU offloading для SD: {offload_err}")
                print("Модель SD останется полностью на CUDA (если возможно) или CPU.")

        # Опционально: компиляция модели для ускорения (требует PyTorch 2.0+)
        # try:
        #    if hasattr(torch, 'compile') and device_sd == "cuda":
        #        pipe_base_global.unet = torch.compile(pipe_base_global.unet, mode="reduce-overhead", fullgraph=True)
        #        print("UNet модели Stable Diffusion скомпилирована для ускорения.")
        # except Exception as compile_err:
        #    print(f"Предупреждение: не удалось скомпилировать UNet: {compile_err}")

    except torch.cuda.OutOfMemoryError as oom_load_sd:
        print(f"!!! OutOfMemoryError (VRAM) при загрузке SD на {device_sd}: {oom_load_sd}")
        if device_sd == "cuda": # Если ошибка на CUDA, пробуем на CPU
            print("Попытка загрузить SD модель только на CPU (fallback из-за OOM на CUDA)...")
            try:
                pipe_base_global = DiffusionPipeline.from_pretrained(
                    sd_model_name,
                    torch_dtype=torch.float32, # CPU работает с float32
                    use_safetensors=True
                )
                # pipe_base_global.to("cpu") # Уже должно быть по умолчанию, если torch_dtype не float16
                print(f"Модель SD '{sd_model_name}' успешно загружена на CPU (fallback). RAM: {get_ram_usage():.2f} MB")
            except Exception as cpu_load_err:
                print(f"КРИТИЧЕСКАЯ ОШИБКА: Не удалось загрузить SD модель даже на CPU: {cpu_load_err}"); traceback.print_exc()
                sys.exit("Не удалось загрузить модель Stable Diffusion.")
        else: # Если OOM произошла уже на CPU
            print(f"КРИТИЧЕСКАЯ ОШИБКА: OutOfMemoryError (RAM) при загрузке SD на CPU: {oom_load_sd}"); traceback.print_exc()
            sys.exit("Не удалось загрузить модель Stable Diffusion из-за нехватки RAM.")
    except Exception as e_sd_load: # Любая другая ошибка загрузки SD
        print(f"КРИТИЧЕСКАЯ ОШИБКА при загрузке Stable Diffusion: {e_sd_load}"); traceback.print_exc()
        sys.exit("Не удалось загрузить модель Stable Diffusion.")

    print(f">>> Настройка окружения и загрузка моделей завершена. RAM: {get_ram_usage():.2f} MB")


def step0_generate_and_save_teaser_script(output_dir_path):
    print("\n" + "="*20 + " ШАГ 0: Генерация Сценария Тизера " + "="*20)

    original_story_path = os.path.join(output_dir_path, CONFIG['PATHS']['ORIGINAL_STORY_FILENAME'])
    try:
        with open(original_story_path, "w", encoding="utf-8") as f:
            f.write(history_text)
        print(f"Исходная история сохранена: {original_story_path}")
    except IOError as e:
        print(f"!!! Ошибка сохранения исходной истории: {e}")
        # Не критично для продолжения, если history_text в памяти

    teaser_script_text = generate_teaser_text_with_gemini(history_text, gemini_model_global)

    teaser_script_path = os.path.join(output_dir_path, CONFIG['PATHS']['TEASER_SCRIPT_FILENAME'])
    try:
        with open(teaser_script_path, "w", encoding="utf-8") as f:
            f.write(teaser_script_text)
        print(f"Сценарий тизера сохранен: {teaser_script_path}")
    except IOError as e:
        print(f"!!! Ошибка сохранения сценария тизера: {e}")
        # Критично, если текст не сохранился и не вернулся из функции
        if not teaser_script_text:
             sys.exit("Не удалось ни сгенерировать, ни сохранить сценарий тизера.")

    return teaser_script_text

def step1_segment_and_save_script(teaser_script_text, output_dir_path):
    print("\n" + "="*20 + " ШАГ 1: Сегментация Сценария на Визуальные Сцены " + "="*20)

    if not teaser_script_text or teaser_script_text == CONFIG['FALLBACKS']['TEASER_SCRIPT']:
        print("ПРЕДУПРЕЖДЕНИЕ: Используется fallback-сценарий или сценарий пуст. Сегментация может быть неоптимальной.")
        # Если сценарий - это fallback, он уже короткий. Можно его не делить или делить на малое число частей.
        if teaser_script_text == CONFIG['FALLBACKS']['TEASER_SCRIPT']:
             # Простая сегментация fallback сценария по предложениям, если их несколько
             sentences = re.split(r'(?<=[.!?…])\s+', teaser_script_text.strip())
             if len(sentences) > 1 and len(sentences) <= CONFIG['VIDEO_ASSEMBLY']['TARGET_VISUAL_SCENES'] + 2:
                 visual_scenes_list = [s.strip() for s in sentences if s.strip()]
                 print(f"Fallback сценарий разделен на {len(visual_scenes_list)} предложений.")
             else: # Если одно предложение или слишком много, используем как есть или делим грубо
                 num_fallback_scenes = min(3, CONFIG['VIDEO_ASSEMBLY']['TARGET_VISUAL_SCENES']) # Не более 3 сцен для короткого fallback
                 words = teaser_script_text.split()
                 words_per_scene = math.ceil(len(words) / num_fallback_scenes) if num_fallback_scenes > 0 else len(words)
                 visual_scenes_list = [" ".join(words[i:i+words_per_scene]) for i in range(0, len(words), words_per_scene)]
                 print(f"Fallback сценарий грубо разделен на {len(visual_scenes_list)} частей.")

        elif not teaser_script_text:
             print("!!! КРИТИЧЕСКАЯ ОШИБКА: Текст сценария для сегментации пуст.")
             sys.exit("Пустой сценарий для сегментации.")
        else: # Другой случай, когда текст есть, но возможно он проблемный
             visual_scenes_list = segment_script_into_visual_scenes(teaser_script_text, gemini_model_global)

    else: # Нормальный случай с сгенерированным сценарием
        visual_scenes_list = segment_script_into_visual_scenes(teaser_script_text, gemini_model_global)

    if not visual_scenes_list: # Если visual_scenes_list все еще пуст
        print("!!! КРИТИЧЕСКАЯ ОШИБКА: Сегментация сценария не дала результатов.")
        sys.exit("Не удалось сегментировать сценарий.")

    visual_scenes_path = os.path.join(output_dir_path, CONFIG['PATHS']['VISUAL_SCENES_FILENAME'])
    try:
        with open(visual_scenes_path, "w", encoding="utf-8") as f:
            json.dump(visual_scenes_list, f, indent=2, ensure_ascii=False)
        print(f"Сегменты сценария ({len(visual_scenes_list)} шт.) сохранены: {visual_scenes_path}")
    except IOError as e:
        print(f"!!! Ошибка сохранения файла сегментов сценария: {e}")
        # Не критично, если visual_scenes_list есть в памяти

    return visual_scenes_list

def step2_generate_and_save_images(visual_scenes_list, output_dir_path):
    print("\n" + "="*20 + " ШАГ 2: Генерация Изображений для Сцен " + "="*20)

    # Собираем префикс стиля и ключевые слова
    style_prefix = CONFIG['IMAGE_GENERATION']['DEFAULT_STYLE_PREFIX']
    additional_keywords = CONFIG['IMAGE_GENERATION']['ADDITIONAL_PROMPT_KEYWORDS']
    full_style_prompt_prefix = f"{style_prefix}, {additional_keywords}" if additional_keywords else style_prefix
    # print(f"DEBUG_STEP2: Полный префикс стиля для промптов: '{full_style_prompt_prefix}'")

    images_output_subdir = os.path.join(output_dir_path, CONFIG['PATHS']['SQUARE_IMAGES_SUBDIR'])
    os.makedirs(images_output_subdir, exist_ok=True)

    generated_image_paths = []
    all_image_prompt_details = []
    successful_scene_indices = [] # Хранит 0-based индексы успешно сгенерированных сцен

    num_scenes_to_process = len(visual_scenes_list)
    if num_scenes_to_process == 0:
        print("!!! ПРЕДУПРЕЖДЕНИЕ: Нет сегментов сценария для генерации изображений.")
        # sys.exit не нужен, так как это может быть обработано дальше (например, созданием заглушки видео)
        # Но если нет изображений, то видео не собрать. Лучше выйти.
        sys.exit("Нет сегментов сценария, генерация изображений невозможна.")


    for i, scene_script_segment in enumerate(visual_scenes_list):
        scene_number_human_readable = i + 1
        print(f"\n--- Обработка визуальной сцены {scene_number_human_readable}/{num_scenes_to_process} ---")
        print(f"  Текст сегмента: '{scene_script_segment[:75]}...'")

        # 1. Получаем смысловое описание от Gemini
        gemini_description = generate_image_prompt_description(scene_script_segment, gemini_model_global)

        # 2. Формируем полный промпт для Stable Diffusion
        # Убираем точки и лишние пробелы из описания Gemini перед добавлением к префиксу
        cleaned_gemini_desc = gemini_description.strip().rstrip('.').strip()

        final_sd_prompt_parts = [full_style_prompt_prefix]
        if cleaned_gemini_desc and cleaned_gemini_desc.lower() != CONFIG['FALLBACKS']['IMAGE_DESCRIPTION'].lower():
            final_sd_prompt_parts.append(cleaned_gemini_desc)
        # Если описание от Gemini было fallback или пустое, не добавляем его, чтобы не дублировать смысл
        # и не делать промпт слишком общим, если fallback описание очень общее.

        final_sd_prompt = ", ".join(filter(None, final_sd_prompt_parts))
        final_sd_prompt = re.sub(r'\s*,\s*', ', ', final_sd_prompt).strip().strip(',') # Очистка запятых
        final_sd_prompt = re.sub(r',{2,}', ',', final_sd_prompt) # Удаление двойных запятых

        print(f"  Финальный промпт для SD: '{final_sd_prompt[:150]}...'")
        all_image_prompt_details.append({
            "scene_index_0based": i, # Сохраняем 0-based индекс
            "scene_text": scene_script_segment,
            "gemini_description": gemini_description,
            "full_sd_prompt": final_sd_prompt
        })

        # 3. Генерируем и сохраняем изображение
        image_filename = f"visual_scene_{scene_number_human_readable:03d}.png" # Используем 1-based для имени файла
        image_output_path = os.path.join(images_output_subdir, image_filename)

        generated_path = generate_stable_diffusion_image(final_sd_prompt, image_output_path, pipe_base_global)

        if generated_path and os.path.exists(generated_path):
            try:
                image_file_size = os.path.getsize(generated_path)
                if image_file_size < CONFIG['IMAGE_GENERATION']['MIN_IMAGE_FILE_SIZE_BYTES_CHECK']:
                    print(f"!!! ПРЕДУПРЕЖДЕНИЕ: Файл изображения {os.path.basename(generated_path)} слишком мал ({image_file_size} байт). Возможно, это заглушка или поврежденный файл.")
                generated_image_paths.append(generated_path)
                successful_scene_indices.append(i) # Сохраняем 0-based индекс успешно обработанной сцены
            except OSError as e_stat:
                print(f"!!! Ошибка доступа к файлу {generated_path} после генерации: {e_stat}.")
        else:
            print(f"!!! ВНИМАНИЕ: Не удалось создать или найти изображение для сцены {scene_number_human_readable}.")
            # Можно добавить создание черной заглушки и здесь, если generate_stable_diffusion_image не вернул путь
            # Но generate_stable_diffusion_image уже должен это делать.

        # Очистка GPU памяти после каждой генерации (если используется CUDA)
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        gc.collect() # Сборка мусора CPU

    # Сохранение информации о промптах
    image_prompts_details_path = os.path.join(output_dir_path, CONFIG['PATHS']['IMAGE_PROMPTS_FILENAME'])
    try:
        with open(image_prompts_details_path, "w", encoding="utf-8") as f:
            json.dump(all_image_prompt_details, f, indent=2, ensure_ascii=False)
        print(f"\nДетали всех промптов для изображений сохранены: {image_prompts_details_path}")
    except IOError as e:
        print(f"!!! Ошибка сохранения файла с деталями промптов: {e}")

    if not generated_image_paths:
        print("!!! КРИТИЧЕСКАЯ ОШИБКА: Не было сгенерировано ни одного изображения. Видео не может быть создано.")
        sys.exit("Генерация изображений провалилась.")

    print(f"\n>>> Генерация изображений завершена. Успешно сгенерировано: {len(generated_image_paths)} изображений.")
    # Возвращаем пути к изображениям, 0-based индексы успешных сцен, и детали промптов
    return generated_image_paths, successful_scene_indices, all_image_prompt_details


def step3_generate_and_save_audio(teaser_script_text, output_dir_path):
    print("\n" + "="*20 + " ШАГ 3: Генерация Аудиодорожки Тизера " + "="*20)

    audio_output_subdir = os.path.join(output_dir_path, CONFIG['PATHS']['AUDIO_SUBDIR'])
    os.makedirs(audio_output_subdir, exist_ok=True)

    full_audio_filename = CONFIG['PATHS']['FULL_AUDIO_FILENAME']
    full_audio_output_path = os.path.join(audio_output_subdir, full_audio_filename)

    # Используем asyncio.run для вызова асинхронной функции TTS
    generated_audio_path = asyncio.run(generate_audio_tts(teaser_script_text, full_audio_output_path))

    if not generated_audio_path or not os.path.exists(generated_audio_path):
        print("!!! КРИТИЧЕСКАЯ ОШИБКА: Аудиодорожка не была сгенерирована или не найдена.")
        sys.exit("Генерация аудио провалилась.")

    print(f"Единая аудиодорожка тизера сохранена: {generated_audio_path}")

    # Определение фактической длительности аудиофайла
    actual_audio_duration_seconds = 0
    audio_clip_for_duration_check = None
    try:
        audio_clip_for_duration_check = AudioFileClip(generated_audio_path)
        actual_audio_duration_seconds = audio_clip_for_duration_check.duration
        print(f"Фактическая длительность аудиодорожки: {actual_audio_duration_seconds:.2f} секунд.")
    except Exception as e_dur:
        print(f"!!! Ошибка при получении длительности аудиофайла: {e_dur}"); traceback.print_exc()
        # Запасной расчет длительности, если MoviePy не смог определить
        avg_wps = CONFIG['MISC']['AVERAGE_WORDS_PER_MINUTE_NARRATION'] / 60.0
        if avg_wps > 0:
            word_count = len(teaser_script_text.split())
            actual_audio_duration_seconds = word_count / avg_wps
            print(f"Используется примерная расчетная длительность аудио: {actual_audio_duration_seconds:.2f} сек.")
        else:
            actual_audio_duration_seconds = CONFIG['GEMINI']['TEASER_TARGET_NARRATION_SECONDS_MAX'] # Крайний случай
            print(f"Не удалось рассчитать длительность, используется максимальная целевая: {actual_audio_duration_seconds:.2f} сек.")
    finally:
        if audio_clip_for_duration_check:
            try:
                audio_clip_for_duration_check.close()
            except Exception:
                pass # Ошибки при закрытии здесь не критичны
        del audio_clip_for_duration_check
        gc.collect()

    return generated_audio_path, actual_audio_duration_seconds

def step5_assemble_video(
    list_of_generated_image_paths,
    list_of_successful_image_indices, # 0-based
    list_of_all_image_prompt_details, # Содержит scene_index_0based
    full_audio_file_path,
    actual_audio_duration_sec,
    base_output_dir_for_run, # Для временных файлов и финального видео
    current_run_timestamp # Для уникальности имен файлов
    ):
    print("\n" + "="*20 + " ШАГ 5: Сборка Видео (через FFMPEG конкатенацию) " + "="*20)
    print(f"RAM на начало сборки видео (step5): {get_ram_usage():.2f} MB")

    num_visual_clips_to_assemble = len(list_of_generated_image_paths)
    if num_visual_clips_to_assemble == 0:
        print("!!! КРИТИЧЕСКАЯ ОШИБКА: Нет изображений для сборки видео.")
        sys.exit("Нет изображений для видео.")

    # Определяем финальную длительность видео: минимум из длины аудио и лимита для Shorts
    target_video_duration_limit = min(actual_audio_duration_sec, CONFIG['VIDEO_ASSEMBLY']['TARGET_SHORT_DURATION_SECONDS'])
    print(f"Целевая итоговая длительность видео (ограничена аудио и лимитом Shorts): {target_video_duration_limit:.2f} секунд.")

    # Расчет длительности каждого визуального клипа
    base_duration_per_visual_clip = target_video_duration_limit / num_visual_clips_to_assemble if num_visual_clips_to_assemble > 0 else 0
    # Гарантируем минимальную длительность клипа
    duration_per_visual_clip = max(base_duration_per_visual_clip, CONFIG['VIDEO_ASSEMBLY']['DEFAULT_MIN_CLIP_DURATION_SECONDS'])
    print(f"Расчетная длительность каждого анимированного визуального клипа: {duration_per_visual_clip:.2f} секунд.")

    # Вероятности для типов движения
    movement_types_available = list(MOVEMENT_PROBABILITIES.keys())
    movement_weights_for_random = list(MOVEMENT_PROBABILITIES.values())

    temp_video_clip_paths_for_ffmpeg = [] # Список путей к временным .mp4 файлам
    temp_files_subdir_path = os.path.join(base_output_dir_for_run, CONFIG['VIDEO_ASSEMBLY']['TEMP_CLIPS_SUBDIR'])
    os.makedirs(temp_files_subdir_path, exist_ok=True)
    print(f"Временные видеоклипы будут сохраняться в: {temp_files_subdir_path}")

    moviepy_objects_created_in_step5 = [] # Для .close() всех объектов MoviePy, созданных здесь

    for i in range(num_visual_clips_to_assemble):
        current_image_path = list_of_generated_image_paths[i]
        # original_0_based_scene_index = list_of_successful_image_indices[i] # Индекс исходного сегмента сценария

        original_scene_index_0based = list_of_successful_image_indices[i]
        prompt_details_for_current_image = next(
            (p for p in list_of_all_image_prompt_details if p.get("scene_index_0based") == original_scene_index_0based),
            None
        )
        scene_description_for_log = f"Сцена {original_scene_index_0based + 1}"
        if prompt_details_for_current_image and prompt_details_for_current_image.get('scene_text'):
            scene_description_for_log = f"'{prompt_details_for_current_image['scene_text'][:60]}...'"

        print(f"\n--- Создание временного видеоклипа {i+1}/{num_visual_clips_to_assemble} для {os.path.basename(current_image_path)} ({scene_description_for_log}) ---")
        print(f"RAM перед созданием MoviePy клипа {i+1}: {get_ram_usage():.2f} MB")

        current_pil_image_obj = None
        moviepy_video_clip_item = None
        try:
            current_pil_image_obj = Image.open(current_image_path).convert("RGB")
            img_width, img_height = current_pil_image_obj.size
            if img_width <= 0 or img_height <= 0:
                raise ValueError(f"Изображение {current_image_path} имеет неверные размеры: {img_width}x{img_height}")

            # Выбор типа движения и параметров зума
            chosen_movement_type = random.choices(movement_types_available, weights=movement_weights_for_random, k=1)[0]
            start_zoom, end_zoom = 1.0, 1.0 # По умолчанию для STATIC

            min_zoom_cfg = CONFIG['ANIMATION']['MIN_ZOOM_FACTOR']
            max_zoom_cfg = CONFIG['ANIMATION']['MAX_ZOOM_FACTOR']

            if chosen_movement_type != MovementType["STATIC"]:
                # Для эффектов панорамирования и зума, начальный зум чуть больше 1.0, конечный - больше начального
                start_zoom = random.uniform(1.0, min_zoom_cfg) # От 1.0 до минимального конфигурированного зума
                end_zoom = random.uniform(min_zoom_cfg, max_zoom_cfg) # От мин. конфиг. до макс. конфиг. зума

            if chosen_movement_type == MovementType["ZOOM_OUT"]:
                start_zoom, end_zoom = end_zoom, start_zoom # Меняем местами для зум-аута
            elif chosen_movement_type in [MovementType["PAN_RIGHT"], MovementType["PAN_LEFT"], MovementType["PAN_UP"], MovementType["PAN_DOWN"]]:
                # Для панорамирования делаем зум менее выраженным, чтобы фокус был на движении
                zoom_range_for_pan = (end_zoom - start_zoom) * CONFIG['ANIMATION']['PAN_ZOOM_AMPLITUDE_REDUCTION']
                end_zoom = start_zoom + zoom_range_for_pan

            print(f"  Движение: {list(MovementType.keys())[list(MovementType.values()).index(chosen_movement_type)]}, Зум: {start_zoom:.2f} -> {end_zoom:.2f}, Длительность клипа: {duration_per_visual_clip:.2f}s")

            # Создаем функцию make_frame
            make_frame_animated_fn = create_advanced_make_frame(
                current_pil_image_obj, img_width, img_height,
                duration_per_visual_clip, chosen_movement_type,
                start_zoom, end_zoom
            )

            # Создаем VideoClip с помощью make_frame
            moviepy_video_clip_item = VideoClip(make_frame=make_frame_animated_fn, duration=duration_per_visual_clip)
            moviepy_objects_created_in_step5.append(moviepy_video_clip_item) # Добавляем для последующей очистки

            # Применение fade эффектов
            fade_duration_cfg = CONFIG['VIDEO_ASSEMBLY']['FADE_DURATION_SECONDS']
            if duration_per_visual_clip > fade_duration_cfg * 2: # Убедимся, что клип достаточно длинный для fade
                moviepy_video_clip_item = moviepy_video_clip_item.fadein(fade_duration_cfg).fadeout(fade_duration_cfg)

            # Сохранение временного клипа БЕЗ АУДИО для FFMPEG
            temp_clip_filename_base = f"temp_clip_{i:03d}.mp4"
            temp_clip_full_path = os.path.join(temp_files_subdir_path, temp_clip_filename_base)

            print(f"  Запись временного видеоклипа {i+1}: {temp_clip_full_path}. RAM: {get_ram_usage():.2f} MB")
            moviepy_video_clip_item.write_videofile(
                temp_clip_full_path,
                codec='libx264', # Быстрый кодек для временных файлов
                audio=False, # Без аудио
                fps=CONFIG['EXPORT']['RENDER_FPS'],
                preset='ultrafast', # Максимально быстро для временных файлов
                threads=max(1, (os.cpu_count() or 2) // 2), # Не все ядра, чтобы система не зависала
                logger=None # 'bar' или None. None для меньшего вывода
            )
            temp_video_clip_paths_for_ffmpeg.append(temp_clip_full_path)
            print(f"  Временный видеоклип {i+1} успешно записан. RAM: {get_ram_usage():.2f} MB")

        except Exception as e_clip_creation:
            print(f"!!! Ошибка при создании или записи временного видеоклипа {i+1}: {e_clip_creation}"); traceback.print_exc()
            # Если клип не создан, его путь не добавится в список, ffmpeg его проигнорирует (если список не пуст)
            # Если это первый клип и он не создался, то могут быть проблемы дальше
        finally:
            # Освобождаем ресурсы MoviePy клипа и PIL изображения как можно скорее
            if moviepy_video_clip_item:
                try: moviepy_video_clip_item.close()
                except Exception: pass # Ошибки закрытия не критичны здесь
            if current_pil_image_obj:
                try: current_pil_image_obj.close()
                except Exception: pass
            del moviepy_video_clip_item # Явное удаление ссылок
            del current_pil_image_obj
            gc.collect() # Сборка мусора после каждого клипа

    if not temp_video_clip_paths_for_ffmpeg:
        print("!!! КРИТИЧЕСКАЯ ОШИБКА: Не было создано ни одного временного видеоклипа для FFMPEG. Сборка невозможна.")
        sys.exit("Провалено создание временных клипов.")

    print(f"\nВсе {len(temp_video_clip_paths_for_ffmpeg)} временных видеоклипов (без аудио) созданы. RAM: {get_ram_usage():.2f} MB")

    # --- Этап 2: Конкатенация временных клипов с помощью FFMPEG ---
    ffmpeg_list_file_path = os.path.join(temp_files_subdir_path, "ffmpeg_concat_list.txt")
    with open(ffmpeg_list_file_path, "w", encoding='utf-8') as f: # encoding для путей с не-ASCII символами
        for clip_path_for_list in temp_video_clip_paths_for_ffmpeg:
            # Для ffmpeg нужны абсолютные пути или пути относительно list файла, и кавычки, если есть пробелы
            # os.path.abspath гарантирует абсолютный путь
            # Используем ' для кавычек, т.к. сам ffmpeg на Windows может предпочитать 'file C:/path/to/file.mp4'
            f.write(f"file '{os.path.abspath(clip_path_for_list)}'\n")

    concatenated_video_no_audio_path = os.path.join(base_output_dir_for_run, f"concatenated_video_no_audio_{current_run_timestamp}.mp4")
    print(f"Объединение временных клипов с помощью FFMPEG в файл: {concatenated_video_no_audio_path}. RAM: {get_ram_usage():.2f} MB")

    # Команда FFMPEG для конкатенации
    ffmpeg_concat_command = [
        "ffmpeg", "-y",                # Перезаписывать выходной файл без запроса
        "-f", "concat",                # Использовать concat демуксер
        "-safe", "0",                  # Разрешить небезопасные пути в файле списка (для абсолютных путей)
        "-i", ffmpeg_list_file_path,   # Входной файл списка
        "-c", "copy",                  # Копировать кодеки (быстро, без перекодирования)
        concatenated_video_no_audio_path # Выходной файл
    ]
    try:
        ffmpeg_process = subprocess.run(ffmpeg_concat_command, check=True, capture_output=True, text=True, encoding='utf-8')
        print("Временные видеоклипы успешно объединены FFMPEG.")
        # print(f"FFMPEG stdout: {ffmpeg_process.stdout}") # Для отладки
    except subprocess.CalledProcessError as e_ffmpeg_concat:
        print(f"!!! КРИТИЧЕСКАЯ ОШИБКА FFMPEG при объединении клипов: {e_ffmpeg_concat}")
        print(f"FFMPEG stderr: {e_ffmpeg_concat.stderr}")
        # Попытка удалить временные файлы перед выходом
        for path_to_del in temp_video_clip_paths_for_ffmpeg: os.remove(path_to_del)
        os.remove(ffmpeg_list_file_path)
        if os.path.exists(concatenated_video_no_audio_path): os.remove(concatenated_video_no_audio_path)
        sys.exit("Ошибка FFMPEG при конкатенации.")
    print(f"RAM после объединения клипов FFMPEG: {get_ram_usage():.2f} MB")

    # --- Этап 3: Добавление аудиодорожки к объединенному видео с помощью MoviePy ---
    final_video_for_export = None
    concatenated_videofileclip_obj = None
    main_audioclip_obj = None

    try:
        print(f"Загрузка объединенного видео ({os.path.basename(concatenated_video_no_audio_path)}) и аудио для финальной сборки.")
        concatenated_videofileclip_obj = VideoFileClip(concatenated_video_no_audio_path)
        moviepy_objects_created_in_step5.append(concatenated_videofileclip_obj)

        main_audioclip_obj = AudioFileClip(full_audio_file_path)
        moviepy_objects_created_in_step5.append(main_audioclip_obj)

        # Длительность объединенного видео может немного отличаться от суммы длительностей из-за ffmpeg
        # Используем target_video_duration_limit как верхнюю границу
        final_clip_duration = min(concatenated_videofileclip_obj.duration, main_audioclip_obj.duration, target_video_duration_limit)
        print(f"Финальная длительность видео после всех корректировок: {final_clip_duration:.2f} секунд.")

        # Обрезаем видео и аудио до финальной длительности, если нужно
        video_to_use_for_final = concatenated_videofileclip_obj.subclip(0, final_clip_duration)
        if video_to_use_for_final != concatenated_videofileclip_obj: # subclip мог создать новый объект
            moviepy_objects_created_in_step5.append(video_to_use_for_final)

        audio_to_use_for_final = main_audioclip_obj.subclip(0, final_clip_duration)
        if audio_to_use_for_final != main_audioclip_obj: # subclip мог создать новый объект
            moviepy_objects_created_in_step5.append(audio_to_use_for_final)

        # Накладываем аудио на видео
        final_video_for_export = video_to_use_for_final.set_audio(audio_to_use_for_final)
        # final_video_for_export НЕ добавляется в moviepy_objects_created_in_step5 здесь,
        # так как он будет возвращен и обработан/закрыт в main после экспорта.
        # Вместо этого, мы добавим его в список в main.

        print(f"Аудиодорожка успешно наложена на объединенное видео. RAM: {get_ram_usage():.2f} MB")

    except Exception as e_final_assembly:
        print(f"!!! КРИТИЧЕСКАЯ ОШИБКА при добавлении аудио или загрузке клипов для финальной сборки: {e_final_assembly}"); traceback.print_exc()
        # Если здесь ошибка, final_video_for_export будет None, и экспорт не произойдет
        final_video_for_export = None # Убедимся, что возвращаем None

    # --- Этап 4: Очистка временных файлов ---
    print("Очистка временных файлов видеоклипов, созданных для FFMPEG...")
    for temp_clip_path in temp_video_clip_paths_for_ffmpeg:
        try:
            os.remove(temp_clip_path)
            # print(f"  Удален временный клип: {temp_clip_path}")
        except OSError as e_remove_temp:
            print(f"  Предупреждение: не удалось удалить временный клип {temp_clip_path}: {e_remove_temp}")
    try:
        os.remove(ffmpeg_list_file_path)
        # print(f"  Удален файл списка FFMPEG: {ffmpeg_list_file_path}")
    except OSError as e_remove_list:
        print(f"  Предупреждение: не удалось удалить файл списка FFMPEG {ffmpeg_list_file_path}: {e_remove_list}")

    return final_video_for_export, moviepy_objects_created_in_step5


def step6_export_video(final_video_clip_obj_to_export, output_dir_path, run_timestamp_str, concatenated_no_audio_path_to_delete):
    if final_video_clip_obj_to_export is None:
        print("!!! ВИДЕО ДЛЯ ЭКСПОРТА ОТСУТСТВУЕТ (final_video_clip_obj_to_export is None). Экспорт отменен.")
        # Попытка удалить промежуточный файл, если он есть и экспорт невозможен
        if concatenated_no_audio_path_to_delete and os.path.exists(concatenated_no_audio_path_to_delete):
            try:
                os.remove(concatenated_no_audio_path_to_delete)
                print(f"Удален промежуточный файл (экспорт отменен): {concatenated_no_audio_path_to_delete}")
            except Exception as e_rem_intermediate:
                print(f"Не удалось удалить промежуточный файл {concatenated_no_audio_path_to_delete}: {e_rem_intermediate}")
        return None

    print("\n" + "="*20 + " ШАГ 6: Экспорт Финального Видео " + "="*20)
    print(f"RAM перед экспортом финального видео: {get_ram_usage():.2f} MB")

    final_video_filename = f"youtube_short_horror_teaser_{run_timestamp_str}.mp4"
    final_video_output_path = os.path.join(output_dir_path, final_video_filename)

    print(f"Экспорт финального видео в: {final_video_output_path}")
    exported_successfully = False
    try:
        render_preset = CONFIG['EXPORT']['RENDER_PRESET']
        render_crf = CONFIG['EXPORT']['RENDER_CRF']
        render_fps = CONFIG['EXPORT']['RENDER_FPS']

        # Определение количества потоков для рендеринга
        cpu_cores = os.cpu_count() or 4 # По умолчанию 4, если os.cpu_count() не возвращает значение
        threads_for_render = max(1, min(cpu_cores - 1, CONFIG['EXPORT']['MAX_THREADS_CPU_CORES']))

        print(f"Параметры экспорта: codec='libx264', audio_codec='aac', fps={render_fps}, preset='{render_preset}', crf={render_crf}, threads={threads_for_render}")

        final_video_clip_obj_to_export.write_videofile(
            final_video_output_path,
            codec='libx264',
            audio_codec='aac',
            fps=render_fps,
            preset=render_preset, # 'medium' хороший баланс, 'ultrafast', 'fast', 'slow', 'slower'
            threads=threads_for_render,
            logger='bar', # 'bar' для прогресс-бара, None для тишины
            ffmpeg_params=[
                "-crf", render_crf,          # Constant Rate Factor (качество, ниже = лучше и больше файл)
                "-pix_fmt", "yuv420p",       # Формат пикселей для совместимости
                "-movflags", "+faststart"    # Для стриминга/веб-проигрывания
            ]
        )
        print(f"\nФинальное видео успешно создано: {final_video_output_path}")
        print(f"RAM после экспорта: {get_ram_usage():.2f} MB")
        exported_successfully = True

        # Проверка длительности через ffprobe (если доступно) или MoviePy
        if os.path.exists(final_video_output_path):
            try:
                # MoviePy может сам проверить длительность
                with VideoFileClip(final_video_output_path) as temp_clip:
                    probed_duration = temp_clip.duration
                print(f"Проверенная длительность финального видео: {probed_duration:.2f} секунд.")
            except Exception as pe:
                print(f"Расчетная длительность из объекта клипа: {final_video_clip_obj_to_export.duration:.2f} сек (ошибка проверки файла: {pe}).")

        return final_video_output_path

    except Exception as export_err:
        print(f"\n!!! КРИТИЧЕСКАЯ ОШИБКА при экспорте финального видео: {export_err}"); traceback.print_exc()
        return None
    finally:
        # Удаление промежуточного объединенного видео БЕЗ ЗВУКА
        # Делаем это здесь, чтобы он был доступен, если финальный экспорт упал, для отладки.
        if concatenated_no_audio_path_to_delete and os.path.exists(concatenated_no_audio_path_to_delete):
            try:
                os.remove(concatenated_no_audio_path_to_delete)
                if exported_successfully:
                    print(f"Удален промежуточный файл (после успешного экспорта): {concatenated_no_audio_path_to_delete}")
                else:
                    print(f"Удален промежуточный файл (экспорт не удался): {concatenated_no_audio_path_to_delete}")
            except Exception as e_rem_final_intermediate:
                print(f"Не удалось удалить промежуточный файл {concatenated_no_audio_path_to_delete}: {e_rem_final_intermediate}")


def cleanup_moviepy_objects(dict_of_objects_to_close):
    print("\n>>> Очистка объектов MoviePy...")
    closed_object_count = 0
    items_to_process_for_cleanup = []

    # Собираем все объекты из словаря (могут быть одиночные или списки)
    for key, value_item in dict_of_objects_to_close.items():
        if isinstance(value_item, list):
            for i, list_member_item in enumerate(value_item):
                if list_member_item: # Убедимся, что элемент списка не None
                    items_to_process_for_cleanup.append((f"{key}[{i}]", list_member_item))
        elif value_item: # Убедимся, что одиночный элемент не None
            items_to_process_for_cleanup.append((key, value_item))

    for obj_name, obj_to_close in items_to_process_for_cleanup:
        if obj_to_close: # Двойная проверка
            try:
                audio_closed_flag_for_log = False
                # Сначала пытаемся закрыть аудио компонент, если он есть
                if hasattr(obj_to_close, 'audio') and obj_to_close.audio:
                    if hasattr(obj_to_close.audio, 'close'):
                        try:
                            obj_to_close.audio.close()
                            audio_closed_flag_for_log = True
                        except Exception as e_audio_close:
                            print(f"    Предупреждение: не удалось закрыть аудио для объекта '{obj_name}': {e_audio_close}")

                # Затем пытаемся закрыть сам видео (или основной) компонент
                if hasattr(obj_to_close, 'close'):
                    obj_to_close.close()
                    log_message = f"  Закрыт объект MoviePy: {obj_name}"
                    if audio_closed_flag_for_log:
                        log_message += " (включая его аудио компонент)"
                    print(log_message)
                    closed_object_count += 1
                elif audio_closed_flag_for_log: # Если закрыли только аудио, но сам объект не имел .close()
                    print(f"  Закрыт аудио компонент для объекта '{obj_name}' (сам объект не имел .close())")
                    closed_object_count +=1 # Считаем как закрытый ресурс

            except AttributeError:
                # Это может случиться, если объект не имеет метода 'close' или 'audio'
                print(f"  Предупреждение: объект '{obj_name}' не имеет ожидаемых методов close/audio для стандартной очистки MoviePy.")
            except Exception as e_generic_close:
                print(f"  Ошибка при попытке закрытия объекта MoviePy '{obj_name}': {e_generic_close}")

    if closed_object_count > 0:
        print(f">>> Попытка закрытия {closed_object_count} объектов/компонентов MoviePy завершена.")
    else:
        print(">>> Не найдено объектов MoviePy для стандартной очистки через .close().")
    gc.collect() # Принудительная сборка мусора

def final_cleanup_gpu_and_globals():
    global pipe_base_global, gemini_model_global
    print(f"\n>>> Финальная очистка глобальных моделей и GPU ресурсов. RAM до: {get_ram_usage():.2f} MB")

    if pipe_base_global is not None:
        try:
            # Если модель на GPU и поддерживает offload, возможно, ее нужно сначала вернуть на CPU перед del
            # Но простого del должно быть достаточно, чтобы PyTorch освободил память.
            print("  Удаление модели Stable Diffusion из памяти...")
            del pipe_base_global
            pipe_base_global = None
            print("  Модель Stable Diffusion удалена.")
        except Exception as e_del_sd:
            print(f"  Ошибка при удалении модели SD: {e_del_sd}")

    if gemini_model_global is not None:
        try:
            # Для Gemini обычно нет явного .close() или .del(), Python сам уберет
            print("  Удаление модели Gemini из памяти...")
            del gemini_model_global
            gemini_model_global = None
            print("  Модель Gemini удалена.")
        except Exception as e_del_gemini:
            print(f"  Ошибка при удалении модели Gemini: {e_del_gemini}")

    if torch.cuda.is_available():
        print("  Очистка кэша CUDA...")
        try:
            torch.cuda.empty_cache()
            print("  Кэш CUDA очищен.")
        except Exception as gpu_clean_err:
            print(f"  Ошибка при очистке кэша CUDA: {gpu_clean_err}")

    gc.collect() # Финальная сборка мусора
    print(f">>> Финальная очистка завершена. RAM после: {get_ram_usage():.2f} MB")

# ==============================================================================
# Основной Управляющий Поток
# ==============================================================================
def main():
    current_run_timestamp_str = datetime.now().strftime("%Y%m%d_%H%M%S")
    base_output_dir_this_run = f"{CONFIG['PATHS']['BASE_OUTPUT_DIR']}_{current_run_timestamp_str}"
    os.makedirs(base_output_dir_this_run, exist_ok=True)

    print(f"=== Запуск Генератора YouTube Shorts Horror Teaser (Версия 5.0.0) ===")
    print(f"Все результаты этого запуска будут сохранены в: {base_output_dir_this_run}")
    print(f"RAM в начале выполнения main(): {get_ram_usage():.2f} MB")

    # Словарь для сбора всех объектов MoviePy, которые нужно будет закрыть
    moviepy_objects_to_cleanup_at_end = {}
    exported_video_final_path = None # Путь к финальному видеофайлу
    concatenated_video_no_audio_temp_path = None # Путь к промежуточному видео без звука

    try:
        setup_environment_and_models() # Загрузка моделей и настройка API

        teaser_script_text = step0_generate_and_save_teaser_script(base_output_dir_this_run)

        visual_scenes_segments = step1_segment_and_save_script(teaser_script_text, base_output_dir_this_run)

        generated_image_paths_list, successful_indices_list, all_prompts_list = \
            step2_generate_and_save_images(visual_scenes_segments, base_output_dir_this_run)

        full_audio_path, actual_audio_duration = \
            step3_generate_and_save_audio(teaser_script_text, base_output_dir_this_run)

        # Путь к промежуточному файлу, который создаст step5 и который нужно будет удалить в step6
        concatenated_video_no_audio_temp_path = os.path.join(
            base_output_dir_this_run,
            f"concatenated_video_no_audio_{current_run_timestamp_str}.mp4"
        )

        # Сборка видео
        final_video_clip_for_export, moviepy_clips_from_step5_assembly = step5_assemble_video(
            generated_image_paths_list,
            successful_indices_list,
            all_prompts_list,
            full_audio_path,
            actual_audio_duration,
            base_output_dir_this_run,
            current_run_timestamp_str
        )
        # Сохраняем объекты, созданные в step5, для последующей очистки
        moviepy_objects_to_cleanup_at_end['step5_intermediate_clips'] = moviepy_clips_from_step5_assembly if moviepy_clips_from_step5_assembly else []

        # Добавляем сам финальный клип в список для очистки, если он был создан
        if final_video_clip_for_export:
            moviepy_objects_to_cleanup_at_end['final_exportable_clip'] = final_video_clip_for_export

        # Экспорт финального видео
        exported_video_final_path = step6_export_video(
            final_video_clip_for_export,
            base_output_dir_this_run,
            current_run_timestamp_str,
            concatenated_video_no_audio_temp_path # Передаем путь для удаления
        )

        if exported_video_final_path and os.path.exists(exported_video_final_path):
            print(f"\n==================== УСПЕХ! ====================")
            print(f"Видео-тизер успешно сгенерирован и сохранен:")
            print(f"  >> {exported_video_final_path}")
            print(f"================================================")
        else:
            print("\n!!!!!!!!!!!!!!!!!!!! ОШИБКА !!!!!!!!!!!!!!!!!!!!")
            print("Не удалось экспортировать финальное видео-тизер.")
            print("Проверьте лог на наличие ошибок на предыдущих шагах.")
            print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")

    except SystemExit as sys_exit_e:
        print(f"\n!!! ВЫПОЛНЕНИЕ ПРЕРВАНО (SystemExit): Код {sys_exit_e.code}")
        print("!!! Проверьте сообщения об ошибках выше для выяснения причины.")
    except Exception as main_flow_exception:
        print(f"\n!!! КРИТИЧЕСКАЯ НЕПЕРЕХВАЧЕННАЯ ОШИБКА В ОСНОВНОМ ПОТОКЕ (main):")
        traceback.print_exc()
    finally:
        print(f"\nRAM перед финальной очисткой в блоке finally: {get_ram_usage():.2f} MB")

        # Очистка объектов MoviePy
        cleanup_moviepy_objects(moviepy_objects_to_cleanup_at_end)

        # Очистка глобальных моделей и GPU
        final_cleanup_gpu_and_globals()

        # Дополнительная очистка временной папки с ffmpeg клипами, если она еще существует
        temp_ffmpeg_clips_dir = os.path.join(base_output_dir_this_run, CONFIG['VIDEO_ASSEMBLY']['TEMP_CLIPS_SUBDIR'])
        if os.path.exists(temp_ffmpeg_clips_dir):
            try:
                # Удаляем файлы, если они там остались (хотя step5 должен был их удалить)
                for f_name in os.listdir(temp_ffmpeg_clips_dir):
                    try: os.remove(os.path.join(temp_ffmpeg_clips_dir, f_name))
                    except: pass
                os.rmdir(temp_ffmpeg_clips_dir) # Пытаемся удалить саму папку
                print(f"Папка временных ffmpeg клипов '{temp_ffmpeg_clips_dir}' удалена.")
            except OSError as e_rm_tempdir:
                print(f"Предупреждение: не удалось полностью удалить папку временных ffmpeg клипов '{temp_ffmpeg_clips_dir}': {e_rm_tempdir}")

        print("\n" + "="*40)
        print("     Генератор YouTube Shorts Завершил Работу     ")
        print("="*40)
        if exported_video_final_path and os.path.exists(exported_video_final_path):
             print(f"Финальный результат (если успешно): {exported_video_final_path}")
        else:
             print(f"Финальный результат НЕ создан. Логи и промежуточные файлы (если есть): {base_output_dir_this_run}")
        print(f"RAM в самом конце выполнения программы: {get_ram_usage():.2f} MB")

if __name__ == '__main__':
    main()
````
